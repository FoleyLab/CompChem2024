```python
import psi4
import numpy as np
from matplotlib import pyplot as plt
```

```python
mol_tmpl = """
C
O 1 **R**
symmetry c2v
"""

r_angstrom = np.array([0.7, 0.8, 0.9, 1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9, 2.0])
options_dict = {
    "basis": "cc-pVTZ",
    "e_convergence": 1e-10,
    "d_convergence": 1e-10,
}

psi4.set_options(options_dict)
#mol = psi4.geometry(mol_tmpl)

#ccsdt_e = psi4.energy('ccsd')

ccsdt_energy = []

for r_val in r_angstrom:
    mol_str = mol_tmpl.replace("**R**", str(r_val))
    mol = psi4.geometry(mol_str)
    ccsdt_e = psi4.energy('ccsd(t)')
    ccsdt_energy.append(ccsdt_e)
```

```python
plt.plot(r_angstrom, ccsdt_energy)
plt.show()
```
