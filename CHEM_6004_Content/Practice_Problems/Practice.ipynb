### 1. You're my density
The Density Matrix provides an alternative way to specify the state of a quantum system, and can be defined
as 
\begin{equation}
{\bf D} = \Psi \Psi^*
\end{equation}
where $\Psi$ is the wavefunction and $\Psi^*$ is its complex conjugate.  Two practically compute the density matrix,
a basis is chosen to expand the wavefunction in, and the densitry matrix elements are defined in terms of the expansion coefficients.  That is, for the following expansion of the wavefunction in the basis of energy eigenstates,
\begin{equation}
\Psi(x) = \sum_i^N c_i \psi_i (x),
\end{equation}
the density matrix elements are given by
\begin{equation}
D_{ij} = c_i c_j^* ,
\end{equation}
where $c_j^*$ is the complex conjugate of the expansion coefficient $c_j$. 

(a) Consider an electron in a box of length $L=10$ atomic units whose wavefunction is given by 
\begin{equation}
\Psi(x) = 1 \: \psi_1(x) + 0 \: \psi_2(x),
\end{equation}
write the $2x2$ Density matrix for this system.

We can also write the Hamiltonian as a matrix (${\bf H}$) in the same basis in which we wrote the Density Matrix, and 
if we do, we can compute the energy expectation value in the following way:
\begin{equation}
\langle E \rangle = {\rm Tr}\left( {\bf H} {\bf D} \right)
\end{equation}
where the trace operationr (${\rm Tr}$) simply sums up the diagonal elements that results from the multiplication
of ${\bf H}$ and ${\bf D}$.

(b) If the elements of the Hamiltonian matrix are given by 
\begin{equation}
H_{ij} = \int_0^L \psi_i^*(x) \hat{H} \psi_j(x) dx,
\end{equation}
compute the energy expectation value of the density matrix you calculated above.

(c) Consider the wavefunction given by 
\begin{equation}
\Psi(x) = \sqrt{\frac{2}{3} } \psi_1(x) + \sqrt{ \frac{1}{3} } \psi_2(x),
\end{equation}
write the corresponding density matrix and compute the energy expectation value using the trace relation given above.

# Who's in Liouville
The analog of the time-dependent Schrodinger equation for the density matrix is called the Liouville equation, and
has the form
\begin{equation}
\frac{d}{dt} {\bf D} = \frac{-i}{\hbar} [{\bf H},{\bf D}],
\end{equation}
where $[{\bf H},{\bf D}] = {\bf H}{\bf D}-{\bf D}{\bf H}$ is known as a commutator of the Hamiltonian and Density matrices.  

Derive the Liouvlle equation from the time-dependent Schrodinger equation and the definition of the Density Matrix given above. 

${\it Hint:}$ The equation of motion of the complex conjugate of the wavefunction can be written as
\begin{equation}
-i\hbar \frac{d}{dt} \Psi^* = \Psi^* \hat{H}.
\end{equation}
